```python
import pandas as pd
from pysankey import sankey
import json

```

```python
drug = 'EMB'

# Define the file path for the JSON file
json_file_path = f'/mnt/storageG1/lwang/Projects/tb_dr_MIC3/individual_models/targets_pred/mic_dic-{drug}.json'

# Read the JSON file into a dictionary
with open(json_file_path, 'r') as file:
    mic_dic = json.load(file)

print(mic_dic)
file_path = f'targets_pred/targets_pred-{drug}.txt'


import pandas as pd
from pysankey import sankey

pd.options.display.max_rows = 8
df = pd.read_csv(
    file_path, sep=' ',  names=['true', 'predicted']
)
df['true'] = df['true'].astype(int)
df = df.astype(str)
df_sorted = df.sort_values(by=['true', 'predicted'])
df_sorted.replace(mic_dic, inplace=True)
# Define a list of 12 distinct colors
distinct_colors = [
    '#1b7ef7', '#12e23f', '#f71b1b', '#f78c1b', '#f3f71b', '#000000',
    '#2AAAAA', '#555555', '#7FFFFF', '#AAAAAA', '#D55554'
]

# Create a color dictionary using the distinct colors
unique_values = pd.concat([df_sorted['true'], df_sorted['predicted']]).unique()
colorDict = dict(zip(unique_values, distinct_colors[:len(unique_values)]))
# colorDict.update({str(k): v for k, v in mic_dic.items()})

sankey(
    df_sorted['true'], df_sorted['predicted'], aspect=20, colorDict=colorDict,
    fontsize=10, figureName=f'targets_pred-{drug}'
)
# Result is in "fruit.png"
```

```python

def draw_sankey(drug, model=''):
    
    # Define the file path for the JSON file
    json_file_path = f'/mnt/storageG1/lwang/Projects/tb_dr_MIC3/individual_models/targets_pred/mic_dic-{drug}.json'

    # Read the JSON file into a dictionary
    with open(json_file_path, 'r') as file:
        mic_dic = json.load(file)

    print(mic_dic)
    file_path = f'targets_pred/targets_pred{model}-{drug}.txt'


    import pandas as pd
    from pysankey import sankey

    pd.options.display.max_rows = 8
    df = pd.read_csv(
        file_path, sep=' ',  names=['true', 'predicted']
    )
    df['true'] = df['true'].astype(int)
    df = df.astype(str)
    df_sorted = df.sort_values(by=['true', 'predicted'])
    df_sorted.replace(mic_dic, inplace=True)
    # Define a list of 12 distinct colors
    distinct_colors = [
        '#1b7ef7', '#12e23f', '#f71b1b', '#f78c1b', '#f3f71b', '#000000',
        '#2AAAAA', '#555555', '#7FFFFF', '#AAAAAA', '#D55554'
    ]
    
    distinct_colors = [
    "#dfa39d",  # Pink
    "#e4d69e",  # Yellow
    "#bce2a6",  # Green
    "#898ab5",  # Violet
    "#a36ed8",  # Violet
    "#6e80d8",  # Blue
    "#d8b3c4",  # Soft Rose
    "#f2cfa5",  # Warm Apricot
    "#aed9c5",  # Mint Cream
    "#c9c4e4",  # Lavender Mist
    "#b3bfe0",  # Powder Blue
    "#e8b9a6",  # Peach Blush
]


    # Create a color dictionary using the distinct colors
    unique_values = pd.concat([df_sorted['true'], df_sorted['predicted']]).unique()
    colorDict = dict(zip(unique_values, distinct_colors[:len(unique_values)]))
    # colorDict.update({str(k): v for k, v in mic_dic.items()})

    sankey(
        df_sorted['true'], df_sorted['predicted'], aspect=200, colorDict=colorDict,
        fontsize=7, figureName=f'targets_pred{model}-{drug}'
    )
    # Result is in "fruit.png"
```

```python
draw_sankey('AMI', '_XGB')
draw_sankey('BDQ', '_XGB')
draw_sankey('RIF', '_XGB')
draw_sankey('EMB', '_XGB')
draw_sankey('AMI')
draw_sankey('BDQ')
draw_sankey('RIF')
draw_sankey('EMB')
```

```python
# Define colors for correct and incorrect predictions
correct_color = '#1b7ef7'  # Blue for correct predictions
incorrect_color = '#f71b1b'  # Red for incorrect predictions

# Create a color dictionary with only two colors
colorDict = {
    'correct': correct_color,
    'incorrect': incorrect_color
}

# Determine if the prediction is correct
df_sorted['is_correct'] = df_sorted['true'] == df_sorted['predicted']
df_sorted['color'] = df_sorted['is_correct'].map({True: 'correct', False: 'incorrect'})

# Use the color column to color the Sankey diagram
sankey(
    df_sorted['true'], df_sorted['predicted'], aspect=20, colorDict=colorDict,
    fontsize=10, figureName=f'targets_pred-{drug}1'
)
```
